```python
import json 
import requests 
import matplotlib.pyplot as plt 
import pandas as pd
import time 
```

```python
def get_json():
    l2=[]
    l=list(range(10))
    for el in l:
        url='https://api.jikan.moe/v3/anime/'+str(el)+'/'
        req=requests.get(url).json()
        time.sleep(4)
        if str(req['status'])=='404':
            el+=1
        else:
            l2.append(req)
            el+=1
    return(l2)

#get_json()
```

```python
def create_dataframe(lst, dataframe):
    i = 0
    for el in lst:
        anime_name=lst[i]['title']
        anime_id=lst[i]['mal_id']
        n_episodes=lst[i]['episodes']
        score=lst[i]['score']
        rank=lst[i]['rank']
        popularity=lst[i]['popularity']
        duration=lst[i]['duration']
        start_airing_time=lst[i]['aired']['from'] 
        end_airing_time=lst[i]['aired']['to']
        i+=1
        dataframe = dataframe.append({'anime_name': anime_name, 'anime_id':anime_id, 'episodes':n_episodes,'score':score,'rank':rank,'popularity':popularity,'duration':duration,'aired_from':start_airing_time,'aired_to':end_airing_time}, ignore_index=True)
    return (dataframe)

#df = pd.DataFrame(columns=['anime_name', 'anime_id', 'episodes','score','rank','popularity','aired_from','aired_to'])
#create_dataframe(l2, df)
```

```python
def dataframe_to_csv(dataframe,filename):
  dataframe.to_csv(filename,index=False)
  return filename
```

```python
def Start():
    animelist = get_json()
    empty_df = pd.DataFrame(columns=['anime_name', 'anime_id', 'episodes','score','rank','popularity','duration','aired_from','aired_to'])
    df = create_dataframe(animelist, empty_df)
    filename = "anime project.csv"
    dataframe_to_csv(df,filename)

Start()
```
